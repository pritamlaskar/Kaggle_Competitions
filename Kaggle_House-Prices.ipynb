```python
# Pandas, Matplotlib, Seaborn, and Numpy
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np

# Classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

# To see all columns and rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
train = pd.read_csv("C:/Users/Pritam Laskar/Documents/Projects/Kaggle Competitions/House Price/train.csv")
test = pd.read_csv("C:/Users/Pritam Laskar/Documents/Projects/Kaggle Competitions/House Price/test.csv")
```

# Data Cleaning

## Train Data

```python
# Making a df for null values

null_table = pd.DataFrame()
null_table['Null Columns'] = (train.isnull().sum()/len(train))*100
null_table['Dtype'] = train.dtypes
null_table
```

```python
# Drop columns with more than 30% null

train.drop(['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'], axis = 1, inplace = True)
```

```python
# Filling null values with mean or mode
train[['LotFrontage', 'MasVnrArea', 'GarageYrBlt']] = train[['LotFrontage', 'MasVnrArea', 'GarageYrBlt']].fillna(train[['LotFrontage', 'MasVnrArea', 'GarageYrBlt']].mean())
train[['MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Electrical', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']] = train[['MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Electrical', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']].fillna(train[['MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Electrical', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']].mode().values[0][0])
```

```python
(train.isnull().sum()/len(train))*100
```

```python
train.describe()
```

### Outlier Removal

```python
train_numcols = train.select_dtypes(include = np.number)

Q3 = train[train_numcols.columns].quantile(.75)
Q1 = train[train_numcols.columns].quantile(.25)
IQR = Q3 - Q1
UL = Q3 + 1.5 * IQR
LL = Q1 - 1.5 * IQR

columns = train_numcols.columns

for train_numcol in columns:
    train[train_numcol] = np.where(train[train_numcol] > UL[train_numcol], UL[train_numcol], np.where(train[train_numcol] < LL[train_numcol], LL[train_numcol], train[train_numcol]))
```

### Outlier Removal

```python
test_numcol = test.select_dtypes(include = np.number)

Q3 = test_numcol.quantile(0.75)
Q1 = test_numcol.quantile(0.25)
IQR = Q3 - Q1
UL = Q3 + 1.5 * IQR
LL = Q1 - 1.5 * IQR

columns = test_numcol.columns

for train_numcol in columns:
    test[train_numcol] = np.where(test[train_numcol] > UL[train_numcol], UL[train_numcol], np.where(test[train_numcol] < LL[train_numcol], LL[train_numcol], test[train_numcol]))
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for i in ['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'SaleType', 'SaleCondition']:
    train[i] = encoder.fit_transform(train[i])
    test[i] = encoder.fit_transform(test[i])
```

```python
train.head()
```

```python
train['Id'] = train['Id'].astype(int)
test['Id'] = test['Id'].astype(int)
```

# Train Test Split

```python
x = train.drop(['SalePrice'], axis = 1)
y = train['SalePrice']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 10)
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size = 0.3, random_state = 10)
```

# ML Models

### Linear Regression

```python
lr = LinearRegression()

lr_model = lr.fit(x_train, y_train)
y_predlr = lr_model.predict(x_val)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_val, y_predlr)
r2_linear = r2_score(y_val, y_predlr)
print('MSE = ', mse)
print('R2 = ', r2_linear)
```

### Decision Tree Regression

```python
dt = DecisionTreeRegressor()

dt_model = dt.fit(x_train, y_train)
y_preddt = dt_model.predict(x_val)
```

```python
mse1 = mean_squared_error(y_preddt, y_val)
r2_scoredt = r2_score(y_preddt, y_val)
print('MSE of DT = ', mse1)
print('R2 Score of DT = ', r2_scoredt)
```

### Random Forest Regression

```python
rf = RandomForestRegressor()

rf_model = rf.fit(x_train, y_train)
y_predrf = rf_model.predict(x_val)
```

```python
mse2 = mean_squared_error(y_val, y_predrf)
r2_scorerf = r2_score(y_val, y_predrf)
print('MSE of RF = ', mse2)
print('R2 Score of RF = ', r2_scorerf)
```

# Submission 

```python
y_pred = rf_model.predict(x_test)
r2_score(y_test, y_pred)
```
